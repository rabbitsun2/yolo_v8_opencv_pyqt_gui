```python
# YOLO v8 training(YOLO v8 nano)
```

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
import torch

#torch.cuda.set_device(0)   # Set to your desired GPU number

model = YOLO("yolov8n.yaml")  # build a new model from scratch
#model = YOLO("yolov8n.pt")  # load a pretrained model (recommended for training)
#model = YOLO("yolov8n.pt", device='gpu')  # load a pretrained model (recommended for training)
model = YOLO("yolov8n.pt")    # Load a pre-trained model
model.to('cuda')

# pre training 
model.train(
    data="data.yaml", 
    imgsz = 416,
    epochs=50, 
    batch=16, 
    workers=0
)

model.export(format='onnx', simplify=True, dynamic=True) #save
```

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
import torch

torch.cuda.set_device(0)   # Set to your desired GPU number

model = YOLO("yolov8x.yaml")  # build a new model from scratch
model = YOLO("yolov8x.pt")
model.to('cuda')

# pre training 
model.train(
    data="data.yaml", 
    imgsz = 416,
    epochs=50, 
    batch=16, 
    workers=0
)

model.export(format='onnx', simplify=True, dynamic=True, imgsz=416) #save
```
